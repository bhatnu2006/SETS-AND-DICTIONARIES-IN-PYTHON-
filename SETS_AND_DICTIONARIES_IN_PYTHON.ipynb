```
#CREATION OF A SET
myset = {"NOVEMBER","DECEMBER","JANUARY"}
print(myset)
```

```
#DUPLICATES NOT ALLOWED
my_set = {"NOVEMBER","DECEMBER","JANUARY","NOVEMBER"}
print(my_set)
```

```
#THE VALUE TRUE AND 1 , FALSE AND 0 ARE CONSIDERED SAME
my_set ={"NOVEMBER","DECEMBER","JANUARY",True,1,2}
print(my_set)
```

```
#A SET WITH STRNGS,INTEGERS AD BOOLEAN VALUES:
set1 = {"abc","34","true"}
print(set1)
```

```
#To CHECK MEMBERSHIP
my_set = {"NOVEMBER","DECEMBER","JANUARY"}
print("NOVEMBER" in my_set)
```

```
myset = {"NOVEMBER","DECEMBER","JANUARY"}
myset.remove("NOVEMBER")
print(myset)
```

```
myset = {"NOVEMBER","DECEMBER","JANUARY"}
myset.add("FEBRUARY")
print(myset)
```

```
#SET OPERATIONS
A = {1,2,3,4,5}
B = {3,4,5,6}

print("Union :",A|B)
print("Intersection :",A&B)
print("Difference :",A-B)
print("Symmetric Difference :",A^B)
```

```
#FROZENSET
x = frozenset({"NOVEMBER","DECEMBER","JANUARY"})
print(x)
print(type(x))
```

```
#ADDING OR REMOVAL OF ELEMENTS ARE NOT ALLOWED IN FROZENSET
X = frozenset({"NOVEMBER","DECEMBER","JANUARY"})
X.remove("DECEMBER")
print(X)
```
