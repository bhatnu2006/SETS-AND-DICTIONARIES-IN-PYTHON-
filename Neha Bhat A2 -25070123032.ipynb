```
#CREATION OF A SET
myset = {"NOVEMBER","DECEMBER","JANUARY"}
print(myset)
```

```
#SET OPERATIONS
A = {1,2,3,4,5}
B = {3,4,5,6}

print("Union :",A|B)
print("Intersection :",A&B)
print("Difference :",A-B)
print("Symmetric Difference :",A^B)
```

```
#FROZENSET
x = frozenset({"NOVEMBER","DECEMBER","JANUARY"})
print(x)
print(type(x))
```

```
#ADDING OR REMOVAL OF ELEMENTS ARE NOT ALLOWED IN FROZENSET
X = frozenset({"NOVEMBER","DECEMBER","JANUARY"})
X.remove("DECEMBER")
print(X)
```

## DICTIONARY


```
# CREATION OF THE DICTIONARY
my_dictionary = {
    "brand" : "Skoda",
    "model" : "kushaq",
    "year" : 2022
}
print(my_dictionary)
```

```
my_dictionary = {
    "brand" : "Skoda",
    "model" : "kushaq",
    "year" : 2022
}
print(my_dictionary["brand"])
```

```
#LATEST VALUER OF THE KEY
my_dictionary = {
    "brand" : "Skoda",
    "model" : "kushaq",
    "year" : 2022,
    "year" : 2024
    }
print(my_dictionary)
```

```
#LENGTH OF THE KEY VALUE PAIR
len(my_dictionary)
```

```
# CHANGING OR ADDING KEY VALUE TO THE DICTIONARY
car={
    "brand":"skoda",
    "model" : "kushaq",
    "year" : 2022,
    "year" : 2024
      }
car["color"]="white"
x =car.values()
print(x)


```

```
# A DICTIONARY STORES STUDENT NAMES AND MARKS . WRITE A PYHTON CODE TO FIND THE STUDENT WHO SCORED THE HIGHEST MARKS
marks={
    "Radha":98,
    "Ghanashyam":97,
    "Ahaan":80,
    "Aditi":99,
    "Shivaanii":79
}

max =max(marks,key =marks.get)
print("highest_score :",max)
print("marks:",marks[max])
```
